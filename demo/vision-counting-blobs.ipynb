# Vision models counting blobs
In this notebook we use a vision-model to count blobs.

```python
from skimage.io import imread, imshow
from bia_bob import bob
bob.__version__
```

```python
image = imread("https://imagej.net/ij/images/blobs.gif")[0]
print(image.shape)
imshow(image);
```

```python
%%bob image
How many blobs do you see in this image?
```

We can also execute `bob` in a loop to see how reproducible results are.

```python
from bia_bob._machinery import Context

for i in range(10):
    # delete chat history
    Context.chat = []
    
    bob("image", "How many blobs do you see in this image?")
```
