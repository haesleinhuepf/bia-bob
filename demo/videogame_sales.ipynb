```python
%%bob
write python code that
* opens the file vgsales.csv using pandas
* if the file doesn't exist, open it from https://zenodo.org/records/5898311/files/vgsales.csv?download=1 
  and save it as vgsales.csv
```

```python
import os
import pandas as pd
import urllib.request

# Check if the file exists
if not os.path.exists('vgsales.csv'):
    # Download the file from the URL
    urllib.request.urlretrieve('https://zenodo.org/records/5898311/files/vgsales.csv?download=1', 'vgsales.csv')

# Open the file using pandas
df = pd.read_csv('vgsales.csv')

# Do something with the dataframe
# For example, print the first 5 rows
print(df.head())
```

```python
%%bob 
group the data by Publisher and 
draw a bar plot of Global_Sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have the data stored in a pandas DataFrame called 'df'

# Group the data by 'Publisher' and calculate the sum of 'Global_Sales'
grouped_data = df.groupby('Publisher')['Global_Sales'].sum()

# Create a bar plot of the grouped data
grouped_data.plot(kind='bar')

# Set the labels and title of the plot
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Global Sales by Publisher')

# Display the plot
plt.show()
```

```python
%%bob
print out the data with the highes Global_Sales
```

```python
import pandas as pd

# Assuming you already have the data stored in a pandas DataFrame called 'df'

# Find the index of the row with the highest 'Global_Sales' value
max_index = df['Global_Sales'].idxmax()

# Get the row with the highest 'Global_Sales' value
max_row = df.loc[max_index]

# Print out the data with the highest 'Global_Sales' value
print(max_row)
```

```python
%%bob
Group the data by Publisher and select the 10 datasets with the highest Global_Sales
```

```python
import pandas as pd

# Assuming you already have the data stored in a pandas DataFrame called 'df'

# Group the data by 'Publisher' and calculate the sum of 'Global_Sales'
grouped_data = df.groupby('Publisher')['Global_Sales'].sum()

# Sort the grouped data in descending order and select the top 10 datasets
top_10_data = grouped_data.nlargest(10)

# Print out the top 10 datasets with the highest 'Global_Sales'
print(top_10_data)
```
