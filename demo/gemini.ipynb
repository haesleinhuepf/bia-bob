```python
from skimage.io import imread
import pyclesperanto as cle
import stackview
import matplotlib.pyplot as plt

# 1. Load the image
image = imread('blobs.tif')

# 2. Segment using Voronoi-Otsu-Labeling
label_image = cle.voronoi_otsu_labeling(image, spot_sigma=2, outline_sigma=2)

# 3. Display the results in a subplot
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
stackview.imshow(image, plot=ax[0])
ax[0].set_title("Original Image")
stackview.imshow(label_image, plot=ax[1])
ax[1].set_title("Label Image")
plt.tight_layout()
plt.show()
```
